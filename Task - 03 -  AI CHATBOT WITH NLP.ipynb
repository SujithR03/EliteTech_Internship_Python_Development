```python
import spacy
import random
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download("punkt")
nltk.download("stopwords")

# Load spaCy's English NLP model
nlp = spacy.load("en_core_web_sm")

# Define a set of predefined responses
responses = {
    "greeting": ["Hello!", "Hi there!", "Hey! How can I help you?"],
    "goodbye": ["Goodbye!", "See you later!", "Take care!"],
    "name": ["I am an NLP chatbot.", "You can call me ChatBot!", "My name is ChatBot!"],
    "default": ["I'm not sure about that.", "Can you rephrase?", "I don't understand."]
}

# Function to clean user input
def clean_text(text):
    tokens = word_tokenize(text.lower())  # Convert to lowercase & tokenize
    words = [word for word in tokens if word.isalnum()]  # Remove punctuation
    words = [word for word in words if word not in stopwords.words("english")]  # Remove stopwords
    return words

# Function to handle user input
def chatbot_response(user_input):
    cleaned_input = clean_text(user_input)

    # Simple rule-based responses
    if any(word in cleaned_input for word in ["hi", "hello", "hey"]):
        return random.choice(responses["greeting"])
    elif any(word in cleaned_input for word in ["bye", "goodbye", "see you"]):
        return random.choice(responses["goodbye"])
    elif "name" in cleaned_input:
        return random.choice(responses["name"])
    
    # Use spaCy for Named Entity Recognition (NER)
    doc = nlp(user_input)
    entities = [ent.text for ent in doc.ents]
    
    if entities:
        return f"I recognized these entities: {', '.join(entities)}"
    
    return random.choice(responses["default"])

# Main loop to chat with the user
print("Chatbot: Hello! Type 'exit' to stop.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    response = chatbot_response(user_input)
    print(f"Chatbot: {response}")
```
